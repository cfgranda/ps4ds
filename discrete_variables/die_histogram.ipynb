```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
# matplotlib.rcParams['text.usetex'] = True

font_size = 15
font_size_ticks = 15

vals = np.arange(1,6+1,1)
histogram = np.array([10, 8, 18, 7, 7, 10])
np.sum(histogram)
pmf = histogram/np.sum(histogram)
bin_edges = np.arange(0.5, 6.5+1, 1)
plt.figure(figsize=(4,3))
plt.bar(vals,histogram, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
plt.xticks(np.arange(1, 6+1, 1))
plt.xticks(fontsize=font_size_ticks) 
plt.ylim([0,20])
plt.yticks([0,5,10,15],fontsize=font_size_ticks)
plt.ylabel("Count",fontsize=font_size,labelpad = 20)
plt.xlabel("Die roll values",fontsize=font_size)

plt.figure(figsize=(4,3))
plt.bar(vals,pmf, width = 0.5, color = "lightgray", edgecolor="black",linewidth = 2)
plt.xticks(np.arange(1, 6+1, 1))
plt.xticks(fontsize=font_size_ticks) 
plt.yticks(fontsize=font_size_ticks)
plt.ylim([0,20/60])
plt.ylabel(r"$p_{X}(a)$",fontsize=font_size,labelpad = 20)
plt.xlabel(r'$a$',fontsize=font_size);
```
